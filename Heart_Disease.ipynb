```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
sns.displot(df['Age'] , kde = True  )
```

```python
sns.displot(df['RestingBP'])
```

## Pie Chart for Categorical

```python
df.groupby('Sex').size().plot(kind = 'pie' , autopct = "%.1f")
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie' , autopct = "%.1f" , title = 'Chest Pain')
```

## Violin Plot

```python
 sns.violinplot(df['Age'])
```

```python
sns.violinplot(x = df['HeartDisease'] , y = df['Sex'])
```

## Joint Plot

```python
sns.jointplot(x = df['Age'] , y = df['MaxHR'] , data = df)
```
